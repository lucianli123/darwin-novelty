```python
from internetarchive import get_item, download, search_items
```

```python
disc=[ "pol", "gen"]
```

```python
for d in disc:
    count=0
    my_file = open(d+"_authors.txt", "r")
    data = my_file.read()
    data = data.split("\n")
    data=set(data)
    my_file.close()
    
    ia_books=[]
    counts=[]
    
    for n in data:
        count+=1
        if count%40==0:
            print(count)
        if count>100000:
            break
        search="creator:(\""+n+"\")"
        res=search_items(search, fields=["title", "creator", "year", "subject"])
        try:
            num=len(res)
        except:
            continue
        if num<1000: 
            try:
                for i in res:

                    try:
                        subj=i["subject"]
                    except:
                        subj=""
                    try:
                        year=int(i["year"])
                    except:
                        year=0
                    try:
                        author=i["creator"]
                    except:
                        author=""
                    if year<1900 and year>1800:
                        ia_books.append([i["identifier"], d, i["title"], author,year])
            except:
                continue
    ia=pd.DataFrame(ia_books)
    ia.columns=["id", "subject", "title", "author", "date"]
    ia.to_csv(d+"_auth_books.csv")
```
